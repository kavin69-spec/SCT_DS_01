```python
import matplotlib.pyplot as plt

# Age and population (in millions)
age_groups = ['0-14', '15-64', '65+']
population = [309, 683, 278]

# Creation of bar chart
plt.figure(figsize=(8, 5))
colors = ['darkgreen', 'darkblue', 'darkred']
plt.bar(age_groups, population, color=colors)

# Labels and title
plt.title("China Population by Age Group (2023 est.)", fontsize=18)
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
for i, v in enumerate(population):
    plt.text(i, v + 10, f"{v}â€¯M", ha='right', fontsize=11)

plt.ylim(0, max(population) * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
